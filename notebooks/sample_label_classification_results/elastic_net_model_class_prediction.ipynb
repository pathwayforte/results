# Train elastic net model to predict binary sample labels and evaluate model with ROC-AUC scores

```python
from random import shuffle, sample
import seaborn as sns
from pathway_forte.constants import *
from pathway_forte.regression import (
    ssgsea_nes_to_df, 
    train_elastic_net_model, 
    get_parameter_values_for_elastic_net, 
)
```

```python
phenotypes = PHENOTYPE_CLASSES.format('brca')

KEGG_SSGSEA_NES = os.path.join(KEGG_SSGSEA,'kegg_brca.tsv')
REACTOME_SSGSEA_NES = os.path.join(REACTOME_SSGSEA,'reactome_brca.tsv')
WIKIPATHWAYS_SSGSEA_NES = os.path.join(WIKIPATHWAYS_SSGSEA,'wikipathways_brca.tsv')
#MERGED_SSGSEA_NES = os.path.join(MERGED_SSGSEA,'gseapy.samples.normalized.es.txt')
```

```python
kegg_x, kegg_y = ssgsea_nes_to_df(KEGG_SSGSEA_NES, phenotypes)
```

```python
shuffled_kegg_y = sample(kegg_y,len(kegg_y))
```

```python
parameter_list = get_parameter_values_for_elastic_net()
#parameter_list = [.1, .5, .7, .9, .95, .99, 1]
```

```python
outer_cv_folds = 5 
inner_cv_folds = 5
```

```python
kegg_scores = train_elastic_net_model(
    x_features=kegg_x, 
    y_labels=kegg_y, 
    outer_cv_splits=outer_cv_folds, 
    inner_cv_splits=inner_cv_folds, 
    hyperparameter_space=parameter_list, 
    scoring_method='roc_auc'
)
```

```python
kegg_scores
```

```python
kegg_shuffled_scores = train_elastic_net_model(
    x_features=kegg_x, 
    y_labels=shuffled_kegg_y, 
    outer_cv_splits=outer_cv_folds, 
    inner_cv_splits=inner_cv_folds, 
    hyperparameter_space=parameter_list, 
    scoring_method='roc_auc'
)
```

```python
kegg_shuffled_scores
```

```python
sns.boxplot(
    y=kegg_scores, 
    showfliers=False, 
    palette="Set2"
).set_title('ROC-AUC scores for {}-fold cross validation from KEGG'.format(outer_cv_folds),fontsize=14)

plt.tight_layout()
plt.ylabel('AUC', fontsize=13)
```

```python
sns.boxplot(
    y=kegg_shuffled_scores, 
    showfliers=False, 
    palette="Set2"
).set_title('Shuffled ROC-AUC scores for {}-fold cross validation from KEGG'.format(outer_cv_folds),fontsize=14)

plt.tight_layout()
plt.ylabel('AUC', fontsize=13)
```
